## K Means Clustering

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = -2 * np.random.rand(100, 2)
X1 = 1 + 2 * np.random.rand(50, 2)

X[50:100, :] = X1
```

```python
plt.scatter(X[ :, 0], X[ :, 1])
plt.show()
```

```python
wcss = []

for i in range(1, 10):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 10), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') 
plt.show()
```

```python
Kmean = KMeans(n_clusters=2)
Kmean.fit(X)
```

```python
Kmean.cluster_centers_
```

```python
plt.scatter(X[ : , 0], X[ : , 1])
plt.scatter(-1.17007275, -0.90218948, s=200, marker='s') # angka dari line diatas
plt.scatter(2.05968207, 2.0378952, s=200, marker='s') # angka dari line diatas
plt.show()
```

```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=150, n_features=2, centers=3, cluster_std=0.5, shuffle=True, random_state=0)
```

```python
plt.scatter(
    X[:, 0], X[:, 1],
    c='white', marker='o',
    edgecolor='black', s=50
)
plt.show()
```

```python
km = KMeans(n_clusters=3)
y_km = km.fit_predict(X)
```

```python
km.cluster_centers_
```

```python
plt.scatter(
    X[y_km == 0, 0], X[y_km == 0, 1],
    s=50, c='lightgreen',
    marker='s', edgecolor='black',
    label='cluster 1'
)
plt.scatter(
    X[y_km == 1, 0], X[y_km == 1, 1],
    s=50, c='orange',
    marker='o', edgecolor='black',
    label='cluster 2'
)

plt.scatter(
    X[y_km == 2, 0], X[y_km == 2, 1],
    s=50, c='lightblue',
    marker='v', edgecolor='black',
    label='cluster 3'
)

plt.scatter(
    km.cluster_centers_[:, 0], km.cluster_centers_[:, 1],
    s=250, marker='*',
    c='red', edgecolor='black',
    label='centroids'
)

plt.legend()
plt.grid()
plt.show()
```

```python
wcss = []

for i in range(1, 10):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 10), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') 
plt.show()
```

# PCA

```python
from sklearn.datasets import load_breast_cancer
```

```python
breast = load_breast_cancer()
```

```python
breast_data = breast.data
```

```python
breast_data.shape
```

```python
breast_label = breast.target
```

```python
breast_label.shape
```

```python
labels = np.reshape(breast_label, (569,1))
```

```python
final_breast_data = np.concatenate([breast_data, labels], axis=1)
```

```python
final_breast_data.shape
```

```python
import pandas as pd
```

```python
breast_dataset = pd.DataFrame(final_breast_data)
```

```python
breast_dataset.head()
```

```python
features = breast.feature_names
```

```python
print(features)
```

```python
features_labels = np.append(features, 'label')
```

```python
print(features_labels)
```

```python
breast_dataset.columns = features_labels
```

```python
breast_dataset.head()
```

```python
breast_dataset['label'].replace(0, 'Benign', inplace=True)
```

```python
breast_dataset['label'].replace(1, 'Malignant', inplace=True)
```

```python
breast_dataset.head()
```

```python
breast_dataset.tail()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
x = breast_dataset.loc[:, features].values
x = StandardScaler().fit_transform(x)
```

```python
x.shape
```

```python
np.std(x)
```

```python
feat_cols = ['feature'+str(i) for i in range(x.shape[1])]
```

```python
print(feat_cols)
```

```python
normalised_breast = pd.DataFrame(x,columns=feat_cols)
normalised_breast.head()
```

```python
from sklearn.decomposition import PCA
pca_breast = PCA(n_components=2)
principalComponents_breast = pca_breast.fit_transform(x)
```

```python
principal_breast_df = pd.DataFrame(data=principalComponents_breast,
                                   columns=['PC 1', 'PC 2'])
```

```python
principal_breast_df.head()
```

```python
print('Variance Ratio: {}'.format(pca_breast.explained_variance_ratio_))
```

# Save and Load Models

```python
# Assigning features and label variables
weather=['Sunny','Sunny','Overcast','Rainy','Rainy','Rainy','Overcast','Sunny','Sunny', 'Rainy','Sunny','Overcast','Overcast','Rainy']
temp=['Hot','Hot','Hot','Mild','Cool','Cool','Cool','Mild','Cool','Mild','Mild','Mild','Hot','Mild']
play=['No','No','Yes','Yes','Yes','No','Yes','No','Yes','Yes','Yes','Yes','Yes','No']

# Import LabelEncoder
from sklearn import preprocessing

#creating labelEncoder
le = preprocessing.LabelEncoder()

# Converting string labels into numbers.
weather_encoded=le.fit_transform(weather)

# Converting string labels into numbers
temp_encoded=le.fit_transform(temp)
label=le.fit_transform(play)

#Combinig weather and temp into single listof tuples
features=list(zip(weather_encoded,temp_encoded))

#Import Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB

#Create a Gaussian Classifier
model = GaussianNB()

# Train the model using the training sets
model.fit(features,label)
```

```python
import pickle

# save the model to disk
filename = 'final_model.sav'
pickle.dump(model, open(filename, 'wb'))
```

```python
load_model = pickle.load(open(filename, 'rb'))
```
