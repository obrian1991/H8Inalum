```python
import matplotlib.pyplot as plt

%matplotlib inline

df_top10.filter(items=['Total', 'cogs', 'gross income']).plot(kind='bar', figsize=(15,10))

#plt.ylabel('Amount of Sales')
#plt.title('Top 10 branch Supermarket')

#plt.show()
```

```python
import matplotlib.pyplot as plt

%matplotlib inline

df_top10['Rating'].plot(kind='bar', figsize=(15,10))

plt.ylabel('Rating of Sales')
plt.title('Top 10 branch Supermarket')

plt.show()
```

```python
payment_method = df_can.groupby(['Payment']).count()
```

```python
payment_method
```

```python
import matplotlib.pyplot as plt

%matplotlib inline

payment_method['Total'].plot(kind='bar', figsize=(15,10))

plt.xlabel('Amount')
plt.ylabel('Paymet Method')
plt.title('Top 10 branch Supermarket')

plt.show()
```
